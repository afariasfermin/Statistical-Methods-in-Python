```python
import numpy as np
import pandas as pd
from os import fspath
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('../storage/real_estate_price_size.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
y = data['price']
x1 = data['size']
```

```python
plt.scatter(x=x1, y=y)
plt.xlabel('size')
plt.ylabel('price')
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
```

```python
results.summary()
```

```python
plt.scatter(x=x1, y=y)
yhat = 223.1787*x1 + 1.019e+05
fig = plt.plot(x1, yhat, lw=2, c='red', label='regression line')
plt.xlabel('Size')
plt.ylabel('Price')
plt.show()
```
